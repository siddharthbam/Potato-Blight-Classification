```python
import tensorflow as tf
from tensorflow.keras import models, layers
import matplotlib.pyplot as plt
```

```python
# creating constants
IMAGE_SIZE = 256
BATCH_SIZE = 32
CHANNELS = 3
EPOCHS = 50
```

```python
dataset = tf.keras.preprocessing.image_dataset_from_directory(
    '/home/archit/Desktop/ds/main/Classification_Potatodisease/datasets',
    shuffle = True,
    image_size = (IMAGE_SIZE, IMAGE_SIZE),
    batch_size = BATCH_SIZE)
```

```python
class_names = dataset.class_names
class_names
```

```python
len(dataset)
# every element in the dataset is a batch of 32 images, so 68*32 = 2176 images in the dataset
```

```python
plt.figure(figsize = (10 , 10))
for image_batch, label_batch in dataset.take(1):
    for i in range(12):
        ax = plt.subplot(3, 4, i+1)
        plt.imshow(image_batch[0].numpy().astype('uint8'))
        plt.axis('off')
        plt.title(class_names[label_batch[0]])
    print('Image batch shape: ', image_batch.shape)
    print('Label batch', label_batch.numpy())
```

```python
def partition_dataset(df ,train_split = 0.8, test_split = 0.1, valid_ds = 0.1, shuffle = True, shuffle_size = 10000):
    if shuffle:
        df = df.shuffle(shuffle_size, seed = 12)
    length = len(df)
    train_l = int(length*train_split)
    test_l = int(length*test_split)
    valid_l = int(length*valid_ds)
    train_df = df.take(train_l)
    valid_df = df.skip(train_l).take(valid_l)
    test_df = df.skip(train_l+valid_l)
    return train_df, valid_df, test_df
  
```

```python
train_df, valid_df, test_df = partition_dataset(dataset)
```

```python
train_df = train_df.cache().shuffle(1000).prefetch(buffer_size = tf.data.AUTOTUNE)
valid_df = valid_df.cache().shuffle(1000).prefetch(buffer_size = tf.data.AUTOTUNE)
test_df = test_df.cache().shuffle(1000).prefetch(buffer_size = tf.data.AUTOTUNE)
```

image preprocessing - creating layers for preprocessing

```python
# scaling the numpy array
resize_and_rescale = tf.keras.Sequential([
    layers.experimental.preprocessing.Resizing(IMAGE_SIZE, IMAGE_SIZE), 
    layers.experimental.preprocessing.Rescaling(1.0/255) # scaling the values to 0-1
])
```

```python
# data augmentation
data_augmentation = tf.keras.Sequential([
    layers.experimental.preprocessing.RandomFlip('horizontal_and_vertical'),
    layers.experimental.preprocessing.RandomRotation(0.2),
])
```

```python
# convolutional neural network
n_classes = 3

model = models.Sequential([
    resize_and_rescale,
    data_augmentation,

    layers.Conv2D(32, (3,3), activation = 'relu', input_shape = (BATCH_SIZE, IMAGE_SIZE, IMAGE_SIZE, CHANNELS)),
    layers.MaxPooling2D(2, 2),
    layers.Conv2D(64, (3,3), activation = 'relu'),
    layers.MaxPooling2D(2, 2),
    layers.Conv2D(64, (3,3), activation = 'relu'),
    layers.MaxPooling2D(2,2),
    layers.Conv2D(64, (3,3), activation = 'relu'),
    layers.MaxPooling2D(2,2),
    layers.Conv2D(64, (3,3), activation = 'relu'),
    layers.MaxPooling2D(2,2),
    layers.Conv2D(64, (3,3), activation = 'relu'),
    layers.MaxPooling2D(2,2),

    layers.Flatten(),
    layers.Dense(64, activation = 'relu'),
    layers.Dense(n_classes, activation = 'softmax')


])

model.build(input_shape = (BATCH_SIZE, IMAGE_SIZE, IMAGE_SIZE, CHANNELS))
```

```python
model.summary()
```

```python
model.compile(
    optimizer = 'adam',
    loss = tf.keras.losses.SparseCategoricalCrossentropy(from_logits = False),
    metrics = ['accuracy']

)
```

```python
history = model.fit(
    train_df, 
    epochs = EPOCHS,
    batch_size=BATCH_SIZE,
    verbose = 1,
    validation_data = valid_df

    )
```
